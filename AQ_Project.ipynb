# #Checking for Stationary test

# DK Fuller Test

# H0=Time series is not Stationary
# Ha=Timeseries is stationary

```python
X = pred_data_int["pm25"].values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

if result[0] < result[4]["5%"]:
    print ("Reject Ho - Time Series is Stationary")
else:
    print ("Failed to Reject Ho - Time Series is Non-Stationary")
```

# Checking for another KPSS Test

```python
from statsmodels.tsa.stattools import kpss
```

# adfuller shows stationary and kpss shows it is not stationary then by applying differencing method we can get stationary

```python
#Differencing
pred_data_int["Differencing method"]=pred_data_int["pm25"]-pred_data_int["pm25"].shift(1)
```

```python
series1=pred_data_int["Differencing method"].values
```

```python
def kpss_test(series1, **kw):    
    statistic, p_value, n_lags, critical_values = kpss(series1, **kw)
    # Format Output
    print(f'KPSS Statistic: {statistic}')
    print(f'p-value: {p_value}')
    print(f'num lags: {n_lags}')
    print('Critial Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
    print(f'Result: The timeseries is {"not " if p_value < 0.05 else ""}stationary')

kpss_test(series1)
```

```python
diff_met=pred_data_int["Differencing method"].dropna()
```
